```python
#filling null values with appropriate data

df.Bathroom = df.Bathroom.fillna(df.Bathroom.median())
df.Parking = df.Parking.fillna(df.Parking.median())
df.Type = df.Type.fillna('Unknown')
df.Per_Sqft = df.Per_Sqft.fillna(df.Per_Sqft.median())
```

```python
#Checking null values again

df.isnull().sum()
```

```python
#saving a clean file in Local System

df.to_csv('magic_bricks_cleandata.csv')
```

```python
#Dtecting Upper limit and Lower limit value
#Step 1 --->>> Find out Q1 and Q3 (25% and 75%)
Q1 = df.Price.quantile(0.25)
Q3 = df.Price.quantile(0.75)
```

```python
Q1,Q3
```

```python
#Step 2 --- >>> Find out the IQR(Inter Quartile Range)

IQR= Q3-Q1
```

```python
IQR
```

```python
#Step 3 --->>> Finding Upper limit and Lower limit 
UL= Q3+1.5*(IQR)
LL= Q1-1.5*(IQR)
```

```python
UL,LL
```

```python
#we can use data filteration method to find out how many of them are outliers

len(df[df.Price>UL])
```

```python
#Step 4 --->>> Replacing the outliers with UL.
#we use np.where method to replace the outliers with UL

df.Price = np.where(df.Price>UL,UL,df.Price)
```
